A string of all characters is given including numbers, 
    case 1 : 
        if the string contains even No. of special characters then:
            numbers in the given has to be printed even followed by odd
            
            example : 
                i/p : t9@42&516
                o/p : 492561
                    explanation : this string consists of 2 special chars so 1st occured even followed by 1st occured odd
    case 2 :
        if the string contains odd No. of special characters then:
            numbers in the given has to be printed odd followed by even
            
            example : 
                i/p : 5u6@25g7#@
                o/p : 56527
                    explanation : this string consists of 3 special chars so 1st occured odd followed by 1st occured even

```python
string = input()
even = []
odd = []
c = 0
k = 0
counter = 0
for i in range(len(string)):
    x = ord(string[i])
    if (x > 47 and x < 58) or (x > 64 and x < 91) or (x > 96 and x < 123):
        if x > 47 and x < 58:
            if x % 2 == 0:
                even.append(chr(x))
            else:
                odd.append(chr(x))
        counter += 1
    else:
        c += 1
e = len(even)
o = len(odd)
ec = 0
oc = 0
res = ""
if(c % 2 == 0):
    k = 1
while(counter >= 0):
    if k == 1:
        if ec < e:
            res = res + even[ec]
        ec+=1
        k = 0
    else:
        if oc < o:
            res = res + odd[oc]
        oc+=1
        k = 1
    counter-=1
print(res)
```

For a given +ve number, Identify the palindrome formed by performing following Operations : 
    * Add number and it's reverse
    * check whether the sum is palindrome or not. If not add the sum and its reverse and repeat the process until a palindrome
      is obtained.
        
    Test Cases : 
        124 ----> o/p : 545
        4 ------> o/p : 8
        195 ----> o/p : 9339

```python
def reverse(num):
    num = num[::-1]
    return int(num)

num = input()
rev = num[::-1]
while(1):
    num = int(num)
    rev = reverse(str(num))
    sm = num + rev
    rev = reverse(str(sm))
    num = sm
    if sm == rev:
        print(sm)
        break
```

Given a string of random numbers, your job is to find the product of the numbers(one is lesser and one is greater) who is already present in the string. 

For instance, a pronic number is a number which is the product of two consecutive integers, that is, a number of the form 
n(n + 1). Like 6 is the pronic number as 2*3 = 6.

```python
import math
s = input()
k = set()
for i in range(len(s)):
    for j in range(i,len(s)):
        k.add(int(s[i:j+1]))
k = sorted(list(k))
lis = []
for i in k:
    for j in range(1,int(math.sqrt(i))+1):
        if((j)*(j+1) == i):
            lis.append(i)
            break
lis.sort()
if(len(lis) == 0):
    print("-1")
else:
    print(lis)
```

We are given a number n. Our task is to check whether the number is circular prime or not.
Circular Prime : A prime number is said to be a circular prime if after any cyclic permutations of the digits, it remains a prime.
    Test Cases :
    
        i/p : 197
        o/p : Circular Prime
        
        i/p : 54
        o/p : Nope
        
    Explanation ----> 197 971 719 (left rotations) are all primes
                ----> 54 45 (left rotaions) are not primes

```python
def rotate(num,rot=1):
    left = num[0:rot]
    right = num[rot:]
    x = right + left
    return x

def prime(num):
    x = int(num)
    c = 0
    for i in range(2,x//2 + 1):
        if x % i == 0:
            c += 1
    if c == 0:
        return 1
    else:
        return 0

num = input()
l = []
for i in range(len(num)):
    num = rotate(num)
    l.append(num)
sm = 0
for i in l:
    sm = sm + prime(i)
if(sm == len(l)):
    print("circular prime")
else:
    print("Nope")
```
